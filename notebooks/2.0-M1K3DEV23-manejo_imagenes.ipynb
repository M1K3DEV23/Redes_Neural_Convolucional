## 1: Importando bibliotecas necesarias


```python
import numpy as np
import matplotlib.pyplot as plt

from skimage import io
```

## 2: Abrimos nuestra imagen de prueba


```python
im = io.imread("../data/Mara.jpeg")
```

```python
plt.imshow(im)
plt.show()
```

## 3: Transformaciones

- Separemos cada canal:


```python
red_channel = im[:, :, 0]
green_channel = im[:, :, 1]
blue_channel = im[:, :, 2]
```

```python
fig, ax = plt.subplots(1, 3, figsize=(16, 9))

ax[0].imshow(red_channel, cmap="gray")
ax[1].imshow(green_channel, cmap="gray")
ax[2].imshow(blue_channel, cmap="gray")

ax[0].set_title("Red Channel")
ax[1].set_title("Green Channel")
ax[2].set_title("Blue Channel")

ax[0].set_xticks([])
ax[0].set_yticks([])
ax[1].set_xticks([])
ax[1].set_yticks([])
ax[2].set_xticks([])
ax[2].set_yticks([])


plt.show()
```

- Observemos el color de cada canal:


```python
# Esta es una cap AUXILIAR llenar de 0s nos permite completar los 3 canales

aux_dim = np.zeros(im.shape[:2])

# Este es rojo porque es (R, AUX, AUX)
red = np.dstack((red_channel, aux_dim, aux_dim)).astype(np.uint8)
# verde (AUX, G, AUX)
green = np.dstack((aux_dim, green_channel, aux_dim)).astype(np.uint8)
# Azul = (AUX, AUX, B)
blue = np.dstack((aux_dim, aux_dim, blue_channel)).astype(np.uint8)
```

```python
fig, ax = plt.subplots(1, 3)

ax[0].imshow(red)
ax[1].imshow(green)
ax[2].imshow(blue)

ax[0].set_title("Red Channel")
ax[0].set_xticks([])
ax[0].set_yticks([])

ax[1].set_title("Green Channel")
ax[1].set_xticks([])
ax[1].set_yticks([])


ax[2].set_title("Blue Channel")
ax[2].set_xticks([])
ax[2].set_yticks([])

plt.show()
```
